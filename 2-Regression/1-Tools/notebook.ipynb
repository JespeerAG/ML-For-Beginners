```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
linnerud_data, linnerud_target = datasets.load_linnerud(return_X_y=True, as_frame=True)
```

```python
label_waist = linnerud_target.loc[:, "Waist"]
label_waist.head()
```

```python
feature_situps = linnerud_data.loc[:, ["Situps"]]
feature_situps.head()
```

```python
label_waist_array = label_waist.to_numpy().reshape(-1, 1)
label_waist_array
```

```python
feature_situps_array = feature_situps.to_numpy().reshape(-1, 1)
feature_situps_array
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(feature_situps_array, label_waist_array, test_size=0.25)
```

```python
X_train
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel("Situps")
plt.ylabel("Waist")
plt.show()
```
